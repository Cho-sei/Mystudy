```python
conditions = list(itertools.product(
    range(3),
    ('palm', 'back'),
    ('left', 'right'),
    (0, 60, 120, 180, 240, 300),
))
df = pd.DataFrame(
    conditions, columns=('view', 'face', 'hand', 'angle'))
df = df.sample(frac=1)
```

```python
df = df.sort_values('view')
df.reset_index(drop=True, inplace=True)
```

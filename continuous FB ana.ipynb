```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
pid = 'Yu'
```

```python
plt.figure(figsize=(10, 10))
for block in range(5):
    with open('result/' + pid + '_FB_ERSP_Day1_b' + str(block) + '.csv', 'r') as f:
        data = list(csv.reader(f, quoting=csv.QUOTE_NONNUMERIC))
        
        ERD_per_list = []
        for FB in data:
            ERD_per_list.append(sum(i < -10 for i in FB) / len(FB))
        plt.subplot(3, 2, block+1)
        plt.plot(ERD_per_list)
        plt.title('day1_block' + str(block+1))
```
